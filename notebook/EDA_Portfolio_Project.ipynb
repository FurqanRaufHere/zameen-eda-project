Removing Duplicates

```python
print("Rows before removing duplicates:", len(df))

df = df.drop_duplicates()

print("Rows after removing duplicates:", len(df))
```

```python
if "URL" in df.columns:
    df = df.drop_duplicates(subset=["URL"], keep="first")
```

```python
import re
import numpy as np

def clean_price(x):
    if pd.isna(x):
        return np.nan

    x = str(x).lower().replace("pkr", "").replace(",", "").replace("\n", " ").strip()

    nums = re.findall(r"[\d\.]+", x)
    if not nums:
        return np.nan

    num = float(nums[0])

    if "crore" in x:
        return num * 10000000   # convert to PKR
    if "lakh" in x:
        return num * 100000
    return num

df["price_clean"] = df["Price"].apply(clean_price)
df[["Price","price_clean"]].head()
```

```python
# Clean Area Column -> Convert units to sqft
def clean_area(x):
    if pd.isna(x):
        return np.nan

    x = str(x).lower()
    nums = re.findall(r"[\d\.]+", x)
    if not nums:
        return np.nan

    num = float(nums[0])

    # conversions
    if "sq. yd" in x or "sq yd" in x or "sq.yd" in x:
        return num * 9          # 1 sq yard = 9 sqft
    if "marla" in x:
        return num * 272.25
    if "kanal" in x:
        return num * 5445
    if "sq ft" in x or "sqft" in x:
        return num              # already sqft

    return num  # fallback

df["area_clean"] = df["Area"].apply(clean_area)
df[["Area","area_clean"]].head()
```

# **Missing Values Treatment**

```python
# Step 3.1 - Identify missing values

missing = df.isnull().sum().sort_values(ascending=False)
print("Missing values per column:\n")
print(missing)
```

```python
# Impute numerical columns

num_cols = ["price_clean", "area_clean"]

for col in num_cols:
    if col in df.columns:
        df[col] = df[col].fillna(df[col].mean())
```

**Justification:** When numerical columns have missing values, replacing with mean keeps the distribution close to original and avoids losing rows.

```python
# Step 3.3 - Impute categorical columns

cat_cols = ["City", "Type", "Purpose", "Location"]

for col in cat_cols:
    if col in df.columns:
        df[col] = df[col].fillna(df[col].mode()[0])
```

**Justification:** Mode is the most frequent category -> replacing missing values with mode prevents creating unrealistic categories and keeps dataset consistent.

```python
# Step 3.4 - Forward-fill for long text fields
if "Description" in df.columns:
    df["Description"] = df["Description"].ffill()
```

**Justification:**
Forward-fill keeps text close to similar listings without introducing fake text.
Safe for scraped datasets where consecutive rows belong to same page/area.

```python
print("\nMissing values after imputation:\n")
print(df.isnull().sum().sort_values(ascending=False))
```

# **Data Cleaning & Consistency**

```python
from fuzzywuzzy import process

# Get unique city names
unique_cities = df["City"].dropna().unique()

# Standardize function
def standardize_city(x):
    if pd.isna(x):
        return x
    best_match, score = process.extractOne(x, unique_cities)
    if score > 85:   # good match threshold
        return best_match
    return x

df["City_clean"] = df["City"].apply(standardize_city)
```

```python
#check
df[["City", "City_clean"]].drop_duplicates().head(20)
```

```python
# Detect & Correct Inconsistent Property Types

df["Type_clean"] = df["Type"].astype(str).str.strip().str.title()

# Map common variations
type_mapping = {
    "Upper Portion": "Portion",
    "Lower Portion": "Portion",
    "Home": "House",
    "Apartment": "Flat",
    "Pent House": "Penthouse",
}

df["Type_clean"] = df["Type_clean"].replace(type_mapping)
```

```python
# Check cleaned types
df["Type_clean"].value_counts()
```

```python
# Calculate price_per_sqft before removing outliers
df['price_per_sqft'] = df['price_clean'] / df['area_clean']

# Remove Outliers (IQR Method)
col = "price_per_sqft"

Q1 = df[col].quantile(0.25)
Q3 = df[col].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

df = df[(df[col] >= lower) & (df[col] <= upper)].copy()
```

```python
df["price_per_sqft"] = df["price_clean"] / df["area_clean"]
```

```python
# Log transformations
df["log_price"] = np.log1p(df["price_clean"])
df["log_area"] = np.log1p(df["area_clean"])
df["log_ppsqft"] = np.log1p(df["price_per_sqft"])
```

```python
# STEP 7 - Insights & Summary
import pandas as pd
import numpy as np
OUT_DIR = '/mnt/data/eda_outputs'
import os
os.makedirs(OUT_DIR, exist_ok=True)

# load cleaned dataframe (if you already have df in memory, skip loading)
DATA_PATH = "/content/drive/MyDrive/Zameen.csv"
df = pd.read_csv(DATA_PATH, encoding='latin1')

# If you already ran earlier cleaning steps in the same notebook, comment above load and use in-memory df.
# For safety, re-apply simple cleaning for price/area and essential features if not present:

import re
def clean_price(x):
    if pd.isna(x): return np.nan
    s = str(x).lower().replace('pkr','').replace('\n',' ').replace(',',' ')
    nums = re.findall(r'[\d\.]+', s)
    if not nums: return np.nan
    val = float(nums[0])
    if 'crore' in s: return val * 1e7
    if 'lakh' in s or 'lac' in s: return val * 1e5
    return val

def clean_area(x):
    if pd.isna(x): return np.nan
    s = str(x).lower().replace(',',' ')
    nums = re.findall(r'[\d\.]+', s)
    if not nums: return np.nan
    val = float(nums[0])
    if 'sq. yd' in s or 'sq yd' in s or 'sq.yd' in s: return val * 9
    if 'marla' in s: return val * 272.25
    if 'kanal' in s: return val * 5445
    return val

if 'price_clean' not in df.columns:
    df['price_clean'] = df['Price'].apply(clean_price)
if 'area_clean' not in df.columns:
    df['area_clean'] = df['Area'].apply(clean_area)

# create price_per_sqft
df['price_per_sqft'] = df['price_clean'] / df['area_clean']
df['price_per_sqft'] = df['price_per_sqft'].replace([np.inf, -np.inf], np.nan)

# Basic cleaning for city/type if missing
if 'City' in df.columns:
    df['City'] = df['City'].astype(str).str.strip().replace({'nan':np.nan})
if 'Type' in df.columns:
    df['Type'] = df['Type'].astype(str).str.strip().replace({'nan':np.nan})

# 1) Top neighborhoods (Location) by avg price_per_sqft (min 5 listings)
area_stats = (df.groupby('Location')
                .agg(count=('price_per_sqft','count'),
                     mean_ppsqft=('price_per_sqft','mean'),
                     median_ppsqft=('price_per_sqft','median'))
                .reset_index())
area_stats = area_stats[area_stats['count'] >= 5].sort_values('mean_ppsqft', ascending=False)

top_10_expensive = area_stats.head(10)
top_10_value = area_stats.sort_values('mean_ppsqft').head(10)

top_10_expensive.to_csv(os.path.join(OUT_DIR, 'top_10_expensive_areas.csv'), index=False)
top_10_value.to_csv(os.path.join(OUT_DIR, 'top_10_value_areas.csv'), index=False)

# 2) Counts and avg ppsqft by Type
type_stats = (df.groupby('Type')
               .agg(count=('price_per_sqft','count'),
                    mean_ppsqft=('price_per_sqft','mean'))
               .reset_index().sort_values('count', ascending=False))
type_stats.to_csv(os.path.join(OUT_DIR, 'type_stats.csv'), index=False)

# 3) City-level summary
city_stats = (df.groupby('City')
               .agg(count=('price_per_sqft','count'),
                    mean_ppsqft=('price_per_sqft','mean'),
                    median_ppsqft=('price_per_sqft','median'))
               .reset_index().sort_values('mean_ppsqft', ascending=False))
city_stats.to_csv(os.path.join(OUT_DIR, 'city_stats.csv'), index=False)

# Print quick console insights
print("Top 10 most expensive areas by mean PKR per sqft (min 5 listings):")
print(top_10_expensive[['Location','count','mean_ppsqft']].head(10).to_string(index=False))
print("\nTop 10 most affordable areas by mean PKR per sqft (min 5 listings):")
print(top_10_value[['Location','count','mean_ppsqft']].head(10).to_string(index=False))

print("\nTop property types by count and mean ppsqft:")
print(type_stats.head(10).to_string(index=False))

# 4) Automated short textual summary (you can paste this into your report)
summary_lines = []
summary_lines.append("Automated EDA Summary:")
summary_lines.append(f"- Dataset total rows: {len(df)}")
summary_lines.append(f"- Listings with valid price_per_sqft: {df['price_per_sqft'].notna().sum()}")
summary_lines.append(f"- City with highest mean price_per_sqft: {city_stats.iloc[0]['City']} ({city_stats.iloc[0]['mean_ppsqft']:.2f} PKR/sqft)")
summary_lines.append(f"- Most common property type: {type_stats.iloc[0]['Type']} ({type_stats.iloc[0]['count']} listings)")
summary_text = "\n".join(summary_lines)

with open(os.path.join(OUT_DIR, 'automated_summary.txt'), 'w') as f:
    f.write(summary_text)

print("\n--- Automated summary ---\n")
print(summary_text)
print("\nSaved CSVs and summary to:", OUT_DIR)
```
